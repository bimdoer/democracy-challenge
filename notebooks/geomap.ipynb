```python
import pandas as pd
from pathlib import Path
from visualisierungen import schweiz_karte_choropleth

# Breite Tabelle: eine Zeile pro «partei»/Aggregat, Kantone als Spalten (…-japroz)
for base in (Path(".."), Path(".")):
    csv_path = base / "data" / "processed" / "df_heatmap_with_positions.csv"
    if csv_path.is_file():
        break
else:
    raise FileNotFoundError("df_heatmap_with_positions.csv nicht gefunden (CWD: notebooks/ oder Projektroot).")
df_wide = pd.read_csv(csv_path)

# Nur Bundesrat-Position (in der Datei heisst die Zeile «br-pos»)
zeile_br = df_wide.loc[df_wide["partei"] == "br-pos"].drop(columns=["partei"]).iloc[0]

df_map = pd.DataFrame({
    "id": ["CH" + str(c).split("-")[0].upper() for c in zeile_br.index],
    "br_pos": pd.to_numeric(zeile_br, errors="coerce"),
})

schweiz_karte_choropleth(
    df_map,
    wert_spalte="br_pos",
    titel="Position: Bundesrat (br-pos)",
    legend_label="Position",
    cmap="coolwarm",
)
```
